## Years At Company Analysis

Before calculating the `Years_At_Company` column, the `Join_Date` column must be
converted from string format into a proper datetime format. This ensures that
date arithmetic can be performed correctly.

### Step 1: Convert Join_Date to datetime
We convert the `Join_Date` column into datetime format so that we can calculate
how long each employee has been with the company.





```python
df["Join_Date"] = pd.to_datetime(df["Join_Date"], errors="coerce")
```

### Step 2: Create Years_At_Company Column
After converting the dates, we calculate the number of years each employee has
worked at the company by subtracting the join date from today's date and
converting the result from days into years.

```python
df["Years_At_Company"] = (pd.Timestamp("today") - df["Join_Date"]).dt.days // 365
```

### Step 3: Visualize the Distribution
Finally, we create a histogram to show the distribution of employee tenure
across the company.

```python
plt.figure(figsize=(5,3))
plt.hist(df["Years_At_Company"], bins=10, edgecolor="black")
plt.title("Years At Company Overview")
plt.xlabel("Years")
plt.ylabel("Total")
plt.show()
```

### Interpretation

Most employees have been with the company for 4–5 years, suggesting a period of
increased hiring during that time. The lowest tenure group is employees with 
6 years at the company, which may indicate that the company was smaller at that 
time or that some long‑term employees have left. Overall, the tenure distribution 
is balanced and does not show extreme turnover or retention issues.

### Outlier Detection (IQR Method)

To identify extreme values in the dataset, the Interquartile Range (IQR) method is applied to all numerical features. The first quartile (Q1) and third quartile (Q3) define the central 50% of the data, while the IQR represents the spread between them. Any value below Q1 − 1.5 × IQR or above Q3 + 1.5 × IQR is considered an outlier. This analysis helps detect unusually low or high values that may indicate data entry errors, rare cases, or meaningful deviations worth further investigation.


```python
num_cols = df.drop(columns=["Phone"]).select_dtypes(include=["int64", "float64"]).columns

outliers = {}

for col in num_cols:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR
    
    outliers[col] = df[(df[col] < lower) | (df[col] > upper)][col]

outliers

```

### Final Interpretation of Outlier Analysis (IQR Method)

Using the IQR method to detect extreme values across all numerical features (excluding the Phone column), no outliers were identified in the dataset. All values for Age, Salary, Years at Company, and Performance Score fall within the normal statistical range defined by Q1 − 1.5 × IQR and Q3 + 1.5 × IQR. This indicates that the dataset is well‑structured, balanced, and free from unusually high or low values that could distort analysis or model performance. Since no outliers were detected, no data points require removal or special handling at this stage.
